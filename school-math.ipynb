```python
import numpy as np
from matplotlib import pyplot as plt

print("START")
```

```python
from math import pi

r = 20

fig, ax = plt.subplots()

x = np.cos(np.arange(-pi, pi, pi/10))
y = np.sin(np.arange(-pi, pi, pi/10))
l = np.arange(-pi, pi, pi/10)

ax.plot(l, x, label='Cos')
ax.plot(l, y, label='Sin')

leg = ax.legend()
plt.show()
```

```python
x = r * x
y = r * y

fig, ax = plt.subplots()
ax.scatter(x, y)
ax.axis('equal')
leg = ax.legend();
```
