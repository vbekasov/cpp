```python
import numpy as np
#from matplotlib import pyplot as plt
%matplotlib notebook

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d  

print("START")
```

```python
from math import pi

r = 20

fig, ax = plt.subplots()

x = np.cos(np.arange(-pi, pi, pi/10))
y = np.sin(np.arange(-pi, pi, pi/10))
l = np.arange(-pi, pi, pi/10)

ax.plot(l, x, label='Cos')
ax.plot(l, y, label='Sin')

leg = ax.legend()
plt.show()
```

```python
r1 = 15
x1 = np.arange(-1*r1, r1+0.01, 0.01)
y1 = np.sqrt(abs(r1*r1 - x1*x1))

x = r * x
y = r * y

fig, ax = plt.subplots()
ax.scatter(x, y)
ax.plot(x1, y1)
ax.plot(x1, -1*y1)
ax.axis('equal')
```

```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D

mpl.rcParams['legend.fontsize'] = 10

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
#ax = fig.gca(projection='3d')
ax.plot(np.zeros(x.shape), x, y, label='vertical')
ax.plot(x, np.zeros(x.shape), y, label='horizontal')
ax.plot(x, y, np.zeros(x.shape), label='desctop')

ax.legend()
ax.set_xlabel(r'X')
ax.set_ylabel(r'Y')
ax.set_zlabel(r'Z')
```

```python
fig, ax = plt.subplots()

l = np.arange(-2*pi, 2*pi, pi/100)
y1 = np.cos(np.arange(-2*pi-pi*(2/3), 2*pi-pi*(2/3), pi/100))
y2 = np.cos(np.arange(-2*pi-pi*(1/3), 2*pi-pi*(1/3), pi/100))
y3 = np.cos(np.arange(-2*pi, 2*pi, pi/100))
y4 = y1 + y2 + y3

ax.grid(True)
ax.plot(l, y1, label='Sin1')
ax.plot(l, y2, label='Sin2')
ax.plot(l, y3, label='Sin3')
ax.plot(l, y4)

leg = ax.legend()
plt.show()

del(y1); del(y2); del(y3); del(y4)
```

```python
mpl.rcParams['legend.fontsize'] = 10

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
#ax = fig.gca(projection='3d')
l = np.arange(-pi, pi, pi/100)

ax.plot(np.zeros(l.shape), np.cos(l), np.sin(l), label='ver1')
ax.plot(np.ones(l.shape), np.cos(l), np.sin(l), label ='ver2')

ax.set_xlabel(r'X')
ax.set_ylabel(r'Y')
ax.set_zlabel(r'Z')
ax.legend()
```

```python
mpl.rcParams['legend.fontsize'] = 10

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
l = np.linspace(pi/2, pi, 10, endpoint=True)

R = 5

ax.plot(np.cos(l), np.sin(np.cos(l))-0.5, np.sin(l))
ax.plot(-1*np.cos(l)+R, -1*np.sin(np.cos(l))-0.5+R, np.sin(l))

ax.set_xlabel(r'X')
ax.set_ylabel(r'Y')
ax.set_zlabel(r'Z')
```

```python
import pandas as pd
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

df1 = pd.DataFrame(data=np.zeros(l.shape)); df1.columns = ['X']
df1['Y'] = np.cos(l); df1['Z'] = np.sin(l)

df2 = pd.DataFrame(data=np.cos(l)); df2.columns = ['X']
df2['Y'] = np.sin(df1['Y']); df2['Z'] = df1['Z']

df1 = pd.concat([df1, df2])
df1 = df1.reset_index()
#df1 = df1.drop(10, axis=0)


pd.options.display.float_format = '{:.4f}'.format
df1.tail()
```

```python
sns.set(style = "darkgrid")
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

x = df1['X']
y = df1['Y']
z = df1['Z']

ax.plot(x, y, z)
plt.show()
```
